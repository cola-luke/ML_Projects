```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
```

```python
breast_cancer_data = load_breast_cancer()
#checking data
print(breast_cancer_data.data[0])
print(breast_cancer_data.feature_names)
print(breast_cancer_data.target, '\n', breast_cancer_data.target_names)
# 0 = malignant

```

```python
#setting x_train, x_test, y_train, y_test
training_data, validation_data, training_labels, validation_labels = train_test_split(breast_cancer_data.data, breast_cancer_data.target, test_size = 0.2
, random_state = 100)

#print(len(training_data) == len(training_labels)) -> True, OK

#setting the classifier

# import warnings filter
from warnings import simplefilter
# ignore all future warnings
simplefilter(action='ignore', category=FutureWarning)

k_list = list(range(1,100))
scores = []
for i in k_list:
    classifier = KNeighborsClassifier(i)
    classifier.fit(training_data, training_labels)
    #checking the score
    scores.append(round(classifier.score(validation_data , validation_labels),4))
    
#Checking the score as a function of K:
plt.plot(k_list , scores, color = 'r')
plt.xlabel('K')
plt.ylabel('Score')
plt.title('How K Affects the Model Performance')
plt.show()



```
